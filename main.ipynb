# Project: Daily To-Do list:
## Overview:
The daily To-Do list is an app that keeps track of your daily tasks. Upon entering the app, you can start adding tasks to your daily list which get reset after 24hrs. Tasks can be added and deleted in between as well. Upon completing all the tasks you will get a congratulations (with sfx) message that will keep you motivated to maintain the streak of completing your daily tasks. You can add previous day's tasks with the click of a single button.

```python
from db import add_task, fetch_tasks, delete_task, mark_task_done, reset_tasks, import_previous_day_tasks
```

```python
def display_menu():
    print("\nDaily To-Do List")
    print("1. Add a Task")
    print("2. View Tasks")
    print("3. Mark a Task as Done")
    print("4. Delete a Task")
    print("5. Reset All Tasks")
    print("6. Import Yesterday's Tasks")
    print("7. Exit")
```

```python
def display_tasks():
    tasks = fetch_tasks()
    if not tasks:
        print("No tasks found!")
    else:
        print("\nYour Tasks:")
        for task in tasks:
            status = "✔️" if task[2] else "❌"
            print(f"{task[0]}. {task[1]} [{status}]")
```

### Main:

```python
# Main app logic
def main():
    while True:
        display_menu()
        choice = input("\nEnter your choice (1-7): ").strip()
        
        if choice == "1":  # Add a Task
            task_title = input("Enter the task description: ").strip()
            if task_title:
                add_task(task_title)
            else:
                print("Task description cannot be empty.")
        
        elif choice == "2":  # View Tasks
            display_tasks()
        
        elif choice == "3":  # Mark Task as Done
            display_tasks()
            try:
                task_id = int(input("\nEnter the task ID to mark as done: ").strip())
                mark_task_done(task_id)
            except ValueError:
                print("Invalid input! Please enter a valid task ID.")
        
        elif choice == "4":  # Delete a Task
            display_tasks()
            try:
                task_id = int(input("\nEnter the task ID to delete: ").strip())
                delete_task(task_id)
            except ValueError:
                print("Invalid input! Please enter a valid task ID.")
        
        elif choice == "5":  # Reset All Tasks
            confirm = input("Are you sure you want to reset all tasks? (yes/no): ").strip().lower()
            if confirm == "yes":
                reset_tasks()
        
        elif choice == "6":  # Import Tasks from yesterday 
            confirm = input("Are you sure you want to import yesterday's tasks? (yes/no): ").strip().lower()
            if confirm == "yes":
                import_previous_day_tasks()
        
        elif choice == "7":  # Exit
            print("Goodbye!")
            break
        
        else:
            print("Invalid choice! Please enter a number between 1 and 6.")

```

### Run:

```python
# Run the app
if __name__ == "__main__":
    main()
```
